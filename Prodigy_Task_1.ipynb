Extraction of top 10 countries having more Population

```python
#filter data for total population
total_population = data1[data1['Series Code'] == 'SP.POP.TOTL']

#sort data based on the total population for 2022
total_population_sorted = total_population.sort_values(by='2022',ascending=False)

#select the top 10 countries
top_10_countries = total_population_sorted.head(10)
print("Top 10 countries having more population:")
print(top_10_countries[['Country Name']])
```

**Bar Plot of Top 10 Countries Having more population**

**In 2022**

```python
#Create bar plot
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x = '2022',y = 'Country Name', data = top_10_countries, palette = 'viridis')
plt.xlabel('Population')
plt.ylabel('Country')
plt.title('Top 10 Countries Having more population (2022)')
plt.show()
```

**In 2016**

```python
#Create bar plot
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x='2016', y = 'Country Name', data = top_10_countries, palette ='viridis')
plt.xlabel('Population')
plt.ylabel('Country')
plt.title('Top Contries having more population (2016)')
plt.show()
```

**Extraction Top 10 Countries with Highest Male and female population**

Top 10 Countries with Highest Male Population

```python
#filter data for male population
male_population = data1[data1['Series Code'] == 'SP.POP.TOTL.MA.IN']

#sort data based on the male population for 2022
male_population_sorted = male_population.sort_values(by='2022',ascending=False)

#select the top 10 countries
top_10_male_countries = male_population_sorted.head(10)
print("Top 10 countries with highest male population:")
print(top_10_male_countries[['Country Name']])

#Create bar plot
plt.figure(figsize =(15,8))
plt.subplot(2,2,1)
sns.barplot(x ='2022', y = 'Country Name', data = top_10_male_countries, palette = 'coolwarm')
plt.xlabel('Population')
plt.ylabel('Country')
plt.title('Top 10 Countries with highest male population (2022)')
plt.show()

```

**Top 10 Countries with Highest Female Population**


```python
#filter data for female population
female_population = data1[data1['Series Code'] == 'SP.POP.TOTL.FE.IN']

#sort data based on the female population for 2022
female_population_sorted = female_population.sort_values(by='2022',ascending=False)

#select the top 10 countries
top_10_female_countries = female_population_sorted.head(10)
print("Top 10 countries with highest female population:")
print(top_10_female_countries[['Country Name']])

#Create barplot
plt.figure(figsize =(15,8))
plt.subplot(2,2,1)
sns.barplot(x ='2022', y = 'Country Name', data = top_10_female_countries, palette = 'coolwarm')
plt.xlabel('Population')
plt.ylabel('Country')
plt.title('Top 10 Countries with highest female population (2022)')
plt.show()
```

**Stacked Bar plot**


**Top  10 Countries with Highest male and Female Population**

```python
#merge male and female population data on 'Country Code'
merge_data=pd.merge(male_population,female_population,on="Country Name",suffixes=("_male","_female"))
```

```python
#merged data
#calculate the total population for each country (male + female)
merge_data["Total Population"] = merge_data["2022_male"] + merge_data ["2022_female"]
merge_data.head()
```

```python
#sort data based on total population in descending order
sorted_data = merge_data.sort_values(by="Total Population",ascending=False)
```

```python
#Select the top 10 countries with the highest total population
top_10_countries = sorted_data.head(10)



```

```python
# Create the stacked bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x="Country Name", y="2022_female", data=top_10_countries, color="purple", label="Female Population")
sns.barplot(x="Country Name", y="2022_male", data=top_10_countries, bottom=top_10_countries["2022_female"], color="pink", label="Male Population")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.show()
```

**Bottom  10 Countries having Male and Female Population (2022)**

```python
#select the top 10 countries with the highest total population
bottom_10_countries = sorted_data.tail(10)
```

```python
#Create the Stacked bar plot
plt.figure(figsize=(12,6))
sns.barplot(x = 'Country Name', y = '2022_female', data = bottom_10_countries, color = 'purple', label = 'Female Population')
sns.barplot(x = 'Country Name', y = '2022_male', data = bottom_10_countries, bottom = bottom_10_countries['2022_female'], color ='pink', label = 'Male Population')
plt.xlabel('Country')
plt.ylabel('Population')
plt.legend()
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```
